```python
%matplotlib inline

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,10))

gs = fig.add_gridspec(3,3)

ax1 = fig.add_subplot(gs[0,0])
x= [1,2,3]
y = [1,2,3]
ax1.bar(x = x, height = y)
ax1.axvline(x = 2, color = 'r')
ax1.plot(x,y, color= 'g')
ax1.scatter(x,y, marker = 'd', color = 'black')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title("Bar Plot")

ax2 = fig.add_subplot(gs[1,1])
ax3 = fig.add_subplot(gs[0,2])
ax4 = fig.add_subplot(gs[1,0])

ax5 = fig.add_subplot(gs[0,1])
ax6 = fig.add_subplot(gs[2,0:3])
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import style
style.use('classic')

fig = plt.figure(figsize=(20,10))

gs = fig.add_gridspec(3,3)

for i in range(3):
    ax1 = fig.add_subplot(gs[i,i])
    x= [1,2,3]
    y = [1,2,3]
    ax1.bar(x = x, height = y)
    ax1.axvline(x = 2, color = 'r')
    ax1.plot(x,y, color= 'g')
    ax1.scatter(x,y, marker = 'd', color = 'black')
    ax1.set_xlabel('X')
    ax1.set_ylabel('Y')
    ax1.set_title("Bar Plot")
    ax1.text(x = 1, y = 2, s= 'PLOT OF')
```
