#Rule for plotting in matplotlib
I will from now one use this method:

fig = plt.figure()
ax1 = fig.add_subplot()
or
gs = fig.add_gridspec(2, 2).

Gridspec is much more powerful but is not common to see.

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (20, 20))
gs = fig.add_gridspec(3, 3)
ax1 = fig.add_subplot(gs[:,0])
labels = descriptors["Descriptor type"].values
heights = descriptors["Number"].values
ax1 = plt.barh(y = labels, width= heights)
plt.show()
```
