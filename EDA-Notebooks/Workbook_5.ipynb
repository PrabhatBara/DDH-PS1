```python
des_class = list(descriptors["Descriptor"])
subset = des_class[1].split(", ")
subset
```

```python
data_smiles = pd.read_csv("../data/smiles.csv")
data_cal_des = pd.read_csv("../data/Calculated_descriptors.csv")

data = data_cal_des[subset]
data.head(5)
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('classic')

fig = plt.figure(figsize = (20, 20))
gs = fig.add_gridspec(3, 3)


for i in range(3):
    
    
    from scipy.stats import norm
    import matplotlib.mlab as mlab
    ax = 'ax' + str(0) + str(i)
    ax = fig.add_subplot(gs[0,i])
    x = data.iloc[:,i].values
    ax = plt.hist(x,bins = 20, color = '#E945f5')
    plt.gca()
    plt.ion()
    plt.title('{} Distribution of Molecules'.format(subset[i]))
    plt.xlabel(subset[i])
    plt.ylabel('Frequency')
    plt.grid(True)
    
    ax = 'ax' + str(1) + str(i)
    ax = fig.add_subplot(gs[1,i])
    x = data.iloc[:,i].values
    ax = plt.boxplot(x)
    plt.gca()
    plt.title('{} Distribution of Molecules'.format(subset[i]))
    plt.xlabel(subset[i])
    plt.ylabel('Frequency')
    plt.grid(True)

    ax = 'ax' + str(2) + str(i)
    ax = fig.add_subplot(gs[2,i])
    x = data.iloc[:,i].values
    ax = plt.boxplot(x)
    plt.gca()
    plt.title('{} Distribution of Molecules'.format(subset[i]))
    plt.xlabel(subset[i])
    plt.ylabel('Frequency')
    plt.grid(True)
```
