## Load the dataset having the Descriptor names

```python
import pandas as pd
descriptors = pd.read_excel("../data/Descriptors.xls")
descriptors.head(10)
```

## Import the libraries

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

## What are the Descriptors we are dealing with?

```python
%matplotlib inline

#STEP 1: create a Figure Object
fig = plt.figure(figsize = (15, 20))

#STEP 2: Create a Grid Space in the Figure
gs = fig.add_gridspec(3, 3)

#STEP 3: Add an axes to the Figure canvas

ax = fig.add_subplot(gs[0:,0])

#STEP 4: Plot on the AXES you've just added
#for this we'll plot a bar graph horizontally
#On the y axis it'll represent categories
#On the x axis it will show the number of values in that category

y = descriptors['Descriptor type'].values
width = descriptors['Number'].values

ax.barh(y= y, width= width, color = '#A14BA9', edgecolor = '#000000')
ax.axvline(x = 50,ymin = 0.01, ymax = 0.99, linestyle = '--', color = '#B60000')
ax.minorticks_on()
ax.grid(True)

#STEP 5: Change axis properties, namely xaxis and y axis
ax.set_xlabel(xlabel = 'Number of Descriptors in each Descriptor Class')
ax.set_ylabel(ylabel = 'Descriptor Class')
#ax.text(x= 0.8,y= 0.9,s = 'Plot Of', transform = ax.transAxes )
#ax.annotate(xy = (0.8,0.9), text = 'Plot Of',transform = ax.transAxes)
```

## Load the dataset of the Calculated Descriptors for each SMILE

```python
data_smiles = pd.read_csv("../data/smiles.csv")
data_cal_des = pd.read_csv("../data/Calculated_descriptors.csv")
```

## Create a list of strings for subsetting the dataset of descriptors

```python
des_class = list(descriptors["Descriptor"])
subset = des_class[3].split(", ")
```

## Subset the dataset

```python
%matplotlib inline

#STEP 1: create a Figure Object
fig = plt.figure(figsize = (9,6 ))

#STEP 2: Create a Grid Space in the Figure
gs = fig.add_gridspec(2, 2)

#STEP 3: Add an axes to the Figure canvas

ax1 = fig.add_subplot(gs[0,0])

#STEP 4: Plot on the AXES you've just added
subset = des_class[3].split(", ")
x = data_cal_des[subset].values
nbins = 10
ax1.hist(x= x, bins= nbins, color = '#A14BA9', edgecolor = '#000000')

ax1.set_xlabel(xlabel = '{}'.format(subset[0]))
ax1.set_ylabel(ylabel = 'Frequency')
ax1.axvline(x = np.mean(x),ymin = 0.01, ymax = 0.99, linestyle = '--', color = '#B60000')
ax1.text(x= 0.7 , y= 0.7 ,
         s = '$\mu$: {} \n $\sigma$: {}'.format(np.round(np.mean(x), 2),np.round(np.std(x), 2)),
         transform = ax1.transAxes)
#STEP 3: Add an axes to the Figure canvas

ax2 = fig.add_subplot(gs[0,1])

#STEP 4: Plot on the AXES you've just added
subset = des_class[4].split(", ")
x = data_cal_des[subset].values
nbins = 10
ax2.hist(x= x, bins= nbins, color = '#A14BA9', edgecolor = '#000000')
ax2.axvline(x = np.mean(x),ymin = 0.01, ymax = 0.99, linestyle = '--', color = '#B60000')
ax2.text(x= 0.7 , y= 0.7 ,
         s = '$\mu$: {} \n $\sigma$: {}'.format(np.round(np.mean(x),2),np.round(np.std(x), 2)),
         transform = ax2.transAxes)


ax2.set_xlabel(xlabel = '{}'.format(subset[0]))
ax2.set_ylabel(ylabel = 'Frequency')



ax3 = fig.add_subplot(gs[1,0])

#STEP 4: Plot on the AXES you've just added
subset = des_class[3].split(", ")
x = data_cal_des[subset].values

ax3.boxplot(x= x)

ax3.set_xlabel(xlabel = '{}'.format(subset[0]))
ax3.set_ylabel(ylabel = 'Frequency')
#ax1.axvline(x = np.mean(x),ymin = 0.01, ymax = 0.99, linestyle = '--', color = '#B60000')
ax3.text(x= 0.7 , y= 0.7 ,
         s = '$\mu$: {} \n $\sigma$: {}'.format(np.round(np.mean(x), 2),np.round(np.std(x), 2)),
         transform = ax3.transAxes)
#STEP 3: Add an axes to the Figure canvas

ax4 = fig.add_subplot(gs[1,1])

#STEP 4: Plot on the AXES you've just added
subset = des_class[4].split(", ")
x = data_cal_des[subset].values

ax4.boxplot(x= x)
#ax4.axvline(x = np.mean(x),ymin = 0.01, ymax = 0.99, linestyle = '--', color = '#B60000')
ax4.text(x= 0.7 , y= 0.7 ,
         s = '$\mu$: {} \n $\sigma$: {}'.format(np.round(np.mean(x),2),np.round(np.std(x), 2)),
         transform = ax4.transAxes)


ax4.set_xlabel(xlabel = '{}'.format(subset[0]))
ax4.set_ylabel(ylabel = 'Frequency')
```
